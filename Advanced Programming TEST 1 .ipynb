```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
import seaborn as sb
sb.set()
import datetime as dt
```

```python
url='https://www.worldometers.info/world-population/uganda-population/'
page= requests.get(url)
soup=BeautifulSoup(page.text, 'lxml')
```

```python
table=soup.find('table', {'class': 'table table-striped table-bordered table-hover table-condensed table-list'})
headers=[]
```

```python
for i in table.find_all('th'):
    title= i.text.strip()
    headers.append(title)
```

```python
df= pd.DataFrame(columns=headers)
```

```python
for row in table.find_all('tr')[1:]:
    data=row.find_all('td')
    row_data=[td.text.strip() for td in data]
    length= len(df)
    df.loc[length]=row_data
```

```python
df.head()
```

```python
df = df.replace(',', '', regex=True)
df = df.replace('%', '', regex=True)
```

```python
df.head()
```

```python
df.drop(columns = [])
```

```python
df.head()
```

```python
df["Year"]=pd.to_numeric(df["Year"],errors='ignore')
df["Population"]=pd.to_numeric(df["Population"],errors='ignore')
df["Yearly Change"]=pd.to_numeric(df["Yearly Change"],errors='ignore')
df["Fertility Rate"]=pd.to_numeric(df["Fertility Rate"],errors='ignore')
df["World Population"]=pd.to_numeric(df["World Population"],errors='ignore')
df["Median Age"]=pd.to_numeric(df["Median Age"],errors='ignore')
df["UgandaGlobal Rank"]=pd.to_numeric(df["UgandaGlobal Rank"],errors='ignore')
df["Country's Share of World Pop"]=pd.to_numeric(df["Country's Share of World Pop"],errors='ignore')
df["Migrants (net)"]=pd.to_numeric(df["Migrants (net)"],errors='ignore')
```
